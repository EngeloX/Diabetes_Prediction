```python
import psycopg2
import pandas as pd
from _py.config import config
```

```python
con = psycopg2.connect(
    dbname = config.user,
    user = config.user,
    host = config.host,
    port = config.port,
    password = config.password
)
con.autocommit = True
cursor = con.cursor()

sql_create_db = f"CREATE DATABASE {config.db}"
try:
    cursor.execute(sql_create_db)
    print('db has been created')
except psycopg2.errors.DuplicateDatabase:
    print('db is already exists')

cursor.close()
con.close()
```

```python
train = pd.read_csv(config.path_data_csv+'/train.csv')
test = pd.read_csv(config.path_data_csv+'/test.csv')
sample_submission = pd.read_csv(config.path_data_csv+'/sample_submission.csv')

tables = {
    "train": train,
    "test": test,
    "sample_submission": sample_submission,
}
```

```python
for table_name, df in tables.items():
    df.to_sql(table_name, config.engine, if_exists="replace", index=False)
    print(f"{table_name} successfully loaded")
```
